```python
#importing the required libraries
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
lr=LabelEncoder()
```

```python
#loading the dataset
dataset=pd.read_csv("general_data.csv")
```

```python
#seeing the first 5 rows of the data
dataset.head()
```

```python
#getting the description of the data
dataset.describe()
```

```python
#getting the info of the data
dataset.info()
```

```python
#converting the variables of type 'object' to type 'int'
dataset['EducationField']= lr.fit_transform(dataset['EducationField']) 
dataset['Attrition']= lr.fit_transform(dataset['Attrition']) 
dataset['BusinessTravel']= lr.fit_transform(dataset['BusinessTravel']) 
dataset['Department']= lr.fit_transform(dataset['Department']) 
dataset['Gender']= lr.fit_transform(dataset['Gender']) 
dataset['JobRole']= lr.fit_transform(dataset['JobRole']) 
dataset['MaritalStatus']= lr.fit_transform(dataset['MaritalStatus']) 
dataset['Over18']= lr.fit_transform(dataset['Over18'])
```

```python
#revisulaising the first 5 columns of the dataset to see if the conversion was successfull
dataset.head()
```

```python
#deleating the columns that are not important and storing the dataset without the column in a new variable (dataset1)
dataset1=dataset.drop(['EmployeeID'], axis = 1) 
dataset1
```

```python
#checking for null values
dataset1.isnull().sum()
```

```python
#dropping the null values 
dataset1=dataset1.dropna(axis=0)
dataset1.isnull().sum()
```

```python
#creating the x variable where all the Independent variables are stored
#creating the y variable where the Dependent variables are stored
x= pd.DataFrame([dataset1['Age'], dataset1['DistanceFromHome'], dataset1['MonthlyIncome'], dataset1['PercentSalaryHike'], dataset1['YearsAtCompany']]).T
y=dataset1["Attrition"]
print(x)
print("-------------------------------------------------------------------")
print(y)
```

```python
#splitting the data into traing set and testing set (20% of the data goes to the testing set)
x_train, x_test,y_train,y_test= train_test_split(x,y, test_size=0.20, random_state=43)
```

```python
#creating the Decision Tree with a depth of 5 nodes
model=tree.DecisionTreeClassifier(max_depth=5)
```

```python
#fitting the model
model.fit(x_train,y_train)
```

```python
#saving the tree's data to a file called "Attrition.dot"
with open("Attrition.dot",'w')as f:
    f= tree.export_graphviz(model, feature_names=['DistanceFromHome','Age','MonthlyIncome','PercentSalaryHike','YearsAtCompany'],out_file=f);
```

```python
#seeing the model's score on training data
model.score(x_train,y_train)
```

```python
#predicting on the unseen data
y_pred=model.predict(x_test)
```

```python
#seeing the model's score on testing data 9unseen data)
metrics.accuracy_score(y_test, y_pred)
```
